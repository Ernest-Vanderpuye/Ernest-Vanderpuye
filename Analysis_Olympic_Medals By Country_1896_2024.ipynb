```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_Olympic_Medalists = pd.read_csv('Olympic_Medalists/all_olympic_medalists_with_continent_clean.csv')
df_Olympic_Medalists.head()
```

```python
#Earliest period in the dataset

df_Olympic_Medalists['year'].min()
```

```python
#latest period in the dataset

df_Olympic_Medalists['year'].max()
```

```python
# Getting information on the data types of each field
df_Olympic_Medalists.info()
```

```python
# From the above with a total count of 20247 games but with 20243 medals we could tell that there were games that had no medals issues.
# First is to find out the full details of those games.
# Then we go ahead to drop them from the list

Events_No_Medals = df_Olympic_Medalists[df_Olympic_Medalists['medal'].isna()]
```

```python
df_Olympic_Medalists[df_Olympic_Medalists['medal'].isna()]
```

```python
# Now drop events (rows) with no medals 

df_Olympic_Medalists_Clean = df_Olympic_Medalists.dropna(subset=['medal'])
```

```python
# Now check if the 'clean' contains any events with no medals

df_Olympic_Medalists_Clean[df_Olympic_Medalists_Clean['medal'].isna()]
```

```python
# So with the olympics they use to allowed 'mixed' team where athlete from different countries can compete against each other
# I will therefore filter out the mixed team into a separte data frame incase we need that for separate analysis

df_Olympic_Medalists_Mixed = df_Olympic_Medalists_Clean[df_Olympic_Medalists_Clean['country'] == 'Mixed Team']
df_Olympic_Medalists_Mixed 
```

```python
# Countries and number of medals per medal class, use the below code to find top 10 and bottom 10

Country_Medal_Count = df_Olympic_Medalists_Clean.groupby(['country', 'medal']).size().reset_index(name='no of medals').sort_values(by='no of medals', ascending=False)
pd.set_option('display.max_rows', None)
print(Country_Medal_Count)
```

```python
# Countries and number of olympic events they have participated in

Country_N0s_Events = df_Olympic_Medalists_Clean.groupby('country')['event_name'].size().reset_index(name='no of events').sort_values(by='no of events', ascending=False)
Country_N0s_Events
```

```python
# Top 10 Countries based on the number of events they have participated in 

Top10_Country_N0s_Events = df_Olympic_Medalists_Clean.groupby('country')['event_name'].size().nlargest(10).reset_index(name='no of events').sort_values(by='no of events', ascending=False)
Top10_Country_N0s_Events
```

```python
# Country and the number of olympic games participated in

Country_N0s_Participation = df_Olympic_Medalists_Clean.groupby('country')['games'].nunique().reset_index(name='no of participation').sort_values(by='no of participation', ascending=False)
pd.set_option('display.max_rows', None)
print(Country_N0s_Participation)
```

```python
# Medals count by continent

df_Olympic_Medalists_Clean.groupby('continent')['medal'].size().reset_index(name='no of medals').sort_values(by='no of medals', ascending=False)
```

```python
# Top 10 in terms of number of olympic games participated in

Top10_Country_N0s_Participation = df_Olympic_Medalists_Clean.groupby('country')['games'].nunique().nlargest(10).reset_index(name='no of participation').sort_values(by='no of participation', ascending=False)
Top10_Country_N0s_Participation
```

```python
# Top 10 count by medals

Top10_Country_N0s_Medals = df_Olympic_Medalists_Clean.groupby('country')['medal'].size().nlargest(10).reset_index(name='total medals').sort_values(by='total medals', ascending=False).head(10)
Top10_Country_N0s_Medals
```

```python
# Country, no of participation, events participated, medals count

df_mergeMedals_Participation =pd.merge(Top10_Country_N0s_Medals, Country_N0s_Participation, on='country', how='left')
Top10_df_Snapshot_CountryPerfm = pd.merge(df_mergeMedals_Participation, Country_N0s_Events, on='country', how='left')
Top10_df_Snapshot_CountryPerfm
```

```python
# Bottom 10 in terms of number of olympic games participated in

df_Olympic_Medalists_Clean.groupby('country')['games'].nunique().nsmallest(10).reset_index(name='no of participation').sort_values(by='no of participation', ascending=False)
```

```python
# Country Medals by distribution of event_gender

Medals_By_Gender = df_Olympic_Medalists_Clean.groupby(['country','event_gender']).size().reset_index(name='medals per gender')
Medals_By_Gender
```

```python
# Top 10 country list

Top_10_countries_List = (df_Olympic_Medalists_Clean.groupby('country')['medal'].size().reset_index(name='total medals').sort_values(by='total medals', ascending=False).head(10))['country'].tolist()
Top_10_countries_List
```

```python
# Filter new data frame to include who countries in top 10

Filtered_Countries_Top10 = Medals_By_Gender[Medals_By_Gender['country'].isin(Top_10_countries_List)]
Filtered_Countries_Top10    
```
